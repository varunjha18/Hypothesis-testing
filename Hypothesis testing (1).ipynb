```python
hdata = pd.read_csv('City_Zhvi_AllHomes.csv')
hdata = hdata.drop(hdata.columns[[0]+list(range(3,51))],axis=1)
hdata2 = pd.DataFrame(hdata[['State','RegionName']])
for year in range(2000,2016):
    hdata2[str(year)+'q1'] = hdata[[str(year)+'-01',str(year)+'-02',str(year)+'-03']].mean(axis=1)
    hdata2[str(year)+'q2'] = hdata[[str(year)+'-04',str(year)+'-05',str(year)+'-06']].mean(axis=1)
    hdata2[str(year)+'q3'] = hdata[[str(year)+'-07',str(year)+'-08',str(year)+'-09']].mean(axis=1)
    hdata2[str(year)+'q4'] = hdata[[str(year)+'-10',str(year)+'-11',str(year)+'-12']].mean(axis=1)
year = 2016    
hdata2[str(year)+'q1'] = hdata[[str(year)+'-01',str(year)+'-02',str(year)+'-03']].mean(axis=1)
hdata2[str(year)+'q2'] = hdata[[str(year)+'-04',str(year)+'-05',str(year)+'-06']].mean(axis=1)
hdata2[str(year)+'q3'] = hdata[[str(year)+'-07',str(year)+'-08']].mean(axis=1)
hdata2 = hdata2.replace({'State':states})
hdata2 = hdata2.set_index(['State','RegionName'])
print(hdata2)
```

```python
unitowns = state_college_df

start = rquarters[0]
hdata = hdata2
bstart = hdata.columns[hdata.columns.get_loc(start) -1]

hdata['ratio'] = hdata[bottom] - hdata[bstart]
hdata = hdata[[bottom,bstart,'ratio']]
hdata = hdata.reset_index()
unitowns_hdata = pd.merge(hdata,unitowns,how='inner',on=['State','RegionName'])
unitowns_hdata['uni'] = True
hdata2 = pd.merge(hdata,unitowns_hdata,how='outer',on=['State','RegionName',bottom,bstart,'ratio'])
hdata2['uni'] = hdata2['uni'].fillna(False)

ut = hdata2[hdata2['uni'] == True]
nut = hdata2[hdata2['uni'] == False]

t,p = ttest_ind(ut['ratio'].dropna(),nut['ratio'].dropna())
    
different = True if p < 0.01 else False

better = "non-university town" if ut['ratio'].mean() < nut['ratio'].mean() else "university town"

print(different, p, better,t)
```

```python
pdata = pd.read_csv('datasets_688730_1207327_death-rates-from-air-pollution.csv')
pdata=pdata.drop(columns={'Indoor air pollution (deaths per 100,000)','Air pollution (total) (deaths per 100,000)','Outdoor ozone pollution (deaths per 100,000)','Code'})
pdata=pdata[pdata['Year']==2017]
pdata=pdata.dropna(axis=0)
pdata.rename(columns={'Outdoor particulate matter (deaths per 100,000)':'Outdoor pollution','Entity':'Region'}, 
                 inplace=True)
pdata=pdata.reset_index()
pdata
```

```python
pcdata = pd.read_csv('region_pollution.csv')
pcdata=pcdata.drop(columns={'Indoor Pollution (deaths per 100000)'})
pcdata.rename(columns={'Outdoor Pollution (deaths per 100000)':'Outdoor pollution'}, 
                 inplace=True)
pcdata
```

```python
poldata = pd.merge(pdata,pcdata,how='inner',on=['Region'])
poldata=poldata.drop(columns={'Year','index'})
poldata.dropna(axis=0)
```

```python
t,p = ttest_ind(poldata['Outdoor pollution_x'],poldata['Outdoor pollution_y'])
```

```python
cdata = pd.read_csv('global_covid19_mortality_rates.csv')
cdata=cdata.drop(columns={'Unnamed: 0','Confirmed','Mortality Ratio','Latitude','Longitude','Country_Code'})
cdata
```

```python
popdata = pd.read_csv('population_by_country_2020.csv')
popdata=popdata[['Country (or dependency)','Population (2020)']]
popdata.rename(columns={'Country (or dependency)':'Country','Population (2020)':'Population'}, 
                 inplace=True)

popdata
```

```python
covdata = pd.merge(popdata,cdata,how='inner',on=['Country'])
covdata
```

```python
covdata['Covid deaths']=(covdata['Deaths']/covdata['Population'])*50000
```

```python
covdata=covdata.drop(columns={'Deaths'})
```

```python
covdata.rename(columns={'Country':'Region'}, 
                 inplace=True)
```
